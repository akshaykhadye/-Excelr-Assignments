```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
RF = RandomForestClassifier()
```

```python
params = {'criterion':['gini','entropy'],'max_depth': np.arange(3, 15),'n_estimators' : np.arange(10,300,50)}
```

```python
gcv = GridSearchCV(RF, params, cv=5)
```

```python
import warnings
warnings.filterwarnings('ignore')

gcv.fit(x,y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
rfc =RandomForestClassifier(n_estimators= 60, max_depth=5, criterion='gini', random_state=100)
```

```python
rfc.fit(x,y)
```

```python
scores = cross_val_score(rfc, x, y, cv=15)
```
