# considering support of 0.025 and confidence of 70%

```python
frequent_itemsets = apriori(books , min_support=0.025 , use_colnames = True )
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets , metric ="confidence" , min_threshold = 0.7) 
```

```python
rules
```

```python
rules_lift = rules[rules.lift > 1]
rules_lift
```

```python
rules_lift.sort_values('lift' , ascending = False)
```

```python
len(rules_lift)
```

```python
plt.scatter(rules_lift['support'],rules_lift['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
